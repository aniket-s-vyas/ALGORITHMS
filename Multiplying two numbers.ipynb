```python
def multiply_classic(n1: int,
                    n2: int) -> int:
    '''
    Function to multiply two numbers
    
    :param n1 : First number
    :param n2 : Second number
    
    return: Multiplication result of the two input positive numbers
    '''
    
    #Convert the input numbers to strings
    n1 = str(n1)
    n2 = str(n2)
    final_result=0
    
    for i in range(len(n2)-1,-1,-1):
        carry=0
        multiplier=n2[i]
        interim_result=""
        
        
        for j in range(len(n1)-1,-1,-1):
            
            current_digit = n1[j]

            result = int(current_digit) * int(multiplier)
            result = result + carry
            
            carry = result//10
            
            if result<=9 or j==0:
                interim_result = str(result) + interim_result
            
            else:
                
                interim_result = str(result%10) + interim_result
        
        interim_result= interim_result + '0'*(len(n2)-1-i)
    
        final_result=final_result+int(interim_result)
    
    return final_result
```

```python
def zeropad(n: str,
            k: int,
           direction :str)->str:
    """
    Function to add k zeros to the input string to the left or right 
    
    :param n: input string
    :param k: number of zeros to be added to the input string
    :param direction: the direction in which the zeros are to be added
    
    :return: string with k zeros added to the input string
    """
    
    if direction.lower()=='left':
        return (n+k*'0')
    elif direction.lower()=='right':
        return (k*'0'+n)
    else:
        raise(Exception("Invalid input value for parameter direction."))
```

```python
def multiply_karatsuba(n1: int,
                      n2: int) -> int :
    '''
    Function to mutiply two input numbers using Karatsuba Algorithm

    :param n1 : First number
    :param n2 : Second number
    
    return: Multiplication result of the two input positive numbers

    '''
    
    # Convert input numbers to strings
    n1=str(n1)
    n2=str(n2)
    
    no_of_digits_n1=len(n1)
    no_of_digits_n2=len(n2)
    
    if no_of_digits_n1==1 and no_of_digits_n2==1:
        return int(n1)*int(n2)
    
    # Convert both the sings to same length by adding zeros
    
    if no_of_digits_n1>no_of_digits_n2 :
        n2= zeropad(n2,no_of_digits_n1-no_of_digits_n2,'right')
    
    elif no_of_digits_n1<no_of_digits_n2 :
        n1= zeropad(n1,no_of_digits_n2-no_of_digits_n1,'right')
    else:
        pass
    
    length_of_inputs=len(n1)
    mid_position=length_of_inputs//2
    
    # Convert the input numbers in the form 
    # n1= a*(10**mid_position) + b
    # n2= c*(10**mid_position) + d
    
    a= n1[:mid_position]
    b= n1[mid_position:]
    c= n2[:mid_position]
    d= n2[mid_position:]
    
    ac=multiply_karatsuba(a,c)
    
    bd=multiply_karatsuba(b,d)

    mid_term= multiply_karatsuba(str(int(a)+int(b)),str(int(c)+int(d)))-ac-bd

    
    no_of_padding_digits=length_of_inputs-mid_position
    
    mutiplication_result= int(zeropad(str(ac),2*no_of_padding_digits,'left'))+ int(zeropad(str(mid_term),no_of_padding_digits,'left'))+int(bd)
    
    return mutiplication_result
```
